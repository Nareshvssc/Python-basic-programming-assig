```python
#Q1
def calc(D):
    C=50
    H=30
    import math
    ans = math.sqrt(2*C*D/H)
    return ans
def ques_1():
    import logging
    C=50
    H=30
    valu_in_str = str(input("enter the value of D in a comma seperate format")).split(",")
    logging.basicConfig(filename="PBA_13.log",level=logging.INFO,format="%(levelname)s%(message)s")
    try:
        a=[int(i) for i in valu_in_str]
        logging.info("executed successfully")
    except Exception as err:
        #print("wrong input", err)
        logging.error("wrong input supplied"+str(err))
    else:
        ans=map(calc,a)
        for i in ans:
            print(i,end=" ")
```
